```python
from google_play_scraper import Sort, reviews

result1, continuation_token = reviews(
    'com.supercell.clashofclans',
    lang='en', 
    country='us', 
    sort=Sort.NEWEST, 
    count=3,
    filter_score_with=5 
)



result2, _ = reviews(
    'com.supercell.clashofclans',
    continuation_token=continuation_token 
)
print(result1, result2)
```

```python
review = result1 + result2


for i in review:
    review_rating = i.get('score')
    review_user = i.get('userName').upper()
    review_text = i.get('content')

    print(f"Rating: {review_rating}")
    print(f"User: {review_user}")
    print(f"Review: {review_text}")
    print()
```
